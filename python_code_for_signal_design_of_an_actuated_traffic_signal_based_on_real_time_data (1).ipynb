Sure, here is the Python code for signal design of an actuated traffic signal based on real time data:

```
import pandas as pd
import numpy as np

# Collect and preprocess real-time data
traffic_data = pd.read_csv("traffic_data.csv") 

# Define initial signal timings
cycle_length = 120  # in seconds
min_green_time = 16  # in seconds
max_green_time = 90  # in seconds
amber_time = 5  # in seconds
all_red_time = 3  # in seconds
phase1_green_time = min_green_time  # initial green time for phase 1
phase2_green_time = min_green_time  # initial green time for phase 2

# Define thresholds for adjusting signal timings based on flow rate
phase1_threshold = 4213  # flow rate threshold for phase 1
phase2_threshold = 2096  # flow rate threshold for phase 2
threshold_change = 30  # amount to increase green time by if threshold is exceeded

# Loop over the real-time data and adjust signal timings accordingly
for index, row in traffic_data.iterrows():
  c=row["flow_rate_1"]- phase1_threshold
  c1=row["flow_rate_2"]- phase2_threshold
  if c>=0 and c1<0:
    phase1_green_time = min(max_green_time, phase1_green_time + threshold_change)
    phase2_green_time = cycle_length - phase1_green_time - amber_time - all_red_time
  elif c>=0 and c1>=0:
    if c>c1:
      phase1_green_time = min(max_green_time, phase1_green_time + threshold_change)
      phase2_green_time = cycle_length - phase1_green_time - amber_time - all_red_time
    else:
      phase2_green_time = min(max_green_time, phase2_green_time + threshold_change)
      phase1_green_time = cycle_length - phase2_green_time - amber_time - all_red_time
  elif c<0 and c1>=0:
    phase2_green_time = min(max_green_time, phase2_green_time + threshold_change)
    phase1_green_time = cycle_length - phase2_green_time - amber_time - all_red_time
  elif c<0 and c1<0:
    phase2_green_time = max(min_green_time, phase2_green_time - threshold_change)
    phase1_green_time = cycle_length - phase2_green_time - amber_time - all_red_time
  
  # Output the signal timings for each timestamp
  print("Timestamp:", row["timestamp"])
  print("Total cycle length:", cycle_length, "seconds")
  print("Phase 1 - Green time:", phase1_green_time, "seconds")
  print("Phase 1 - Amber time:", amber_time, "seconds")
  print("Phase 1 - Red time:", cycle_length - phase1_green_time - amber_time - all_red_time, "seconds")
  print("Phase 2 - Green time:", phase2_green_time, "seconds")
  print("Phase 2 - Amber time:", amber_time, "seconds")
  print("Phase 2 - Red time:", cycle_length - phase2_green_time - amber_time - all_red_time, "seconds")
  print("All Red time:", all_red_time, "seconds")
```
